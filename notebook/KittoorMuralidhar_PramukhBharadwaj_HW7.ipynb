i. Research exact match and hamming score/ loss methods for evaluating multi- label classification and use them in evaluating the classifiers in this problem.

There are multiple example-based metrics to be used for multi label classification.

##### Exact Match Ratio

One trivial way around would just to ignore partially correct (consider them incorrect) and extend the accuracy used in single label case for multi-label prediction.

Exact Match Ratio, MR=1/n∑(i=1 to N) I(yi=yi^)
where I is the indicator function. Clearly, a disadvantage of this measure is that it does not distinguish between complete incorrect and partially correct which might be considered harsh.


##### Hamming Score

Hamming score for each instance is defined as the proportion of the predicted correct labels to the total number (predicted and actual) of labels for that instance. Overall accuracy is the average across all instances. 

Accuracy=1/n∑(i=1 to n)|yi∩yi^|/|yi∪yi^|

##### Hamming Loss

It reports how many times on average, the relevance of an example to a class label is incorrectly predicted. Therefore, hamming loss takes into account the prediction error (an incorrect label is predicted) and missing error (a relevant label not predicted), normalized over total number of classes and total number of examples.

Hamming Loss=1/nL∑(i=1 to n)∑(j=1 to L) I(yji≠ŷ ji)
where I is the indicator function. Ideally, we would expect the hamming loss to be 0, which would imply no error; practically the smaller the value of hamming loss, the better the performance of the learning algorithm.

Doing Binary Relevance

```python
def getMetrics(y_true, y_pred):
    exact_match_ratio = 0
    for true_label, pred_label in zip(y_true.values, y_pred.values):
        matched = (true_label == pred_label)
        if sum(matched) == y_true.shape[1]:
            exact_match_ratio += 1;
    exact_match_ratio /= y_true.shape[0]
    
    miss_classified = 0
    for true_label, pred_label in zip(y_true.values, y_pred.values):
        miss_match = (true_label != pred_label)
        miss_classified += np.sum(miss_match)
    hamming_loss = miss_classified / (y_true.shape[0] * y_true.shape[1])
        
    return exact_match_ratio, hamming_loss
    
```

```python
tuned_parameters = [
    { "C": [0.00001, 0.0001, 0.001, 0.01, 0.1, 1, 10, 100, 1000], 'gamma': [1, 0.1, 0.01, 0.001, 0.0001]},
]
kf = StratifiedKFold(10, shuffle = True, random_state = 7)

y_pred_train_df = pd.DataFrame(columns=y_train.columns)
y_pred_test_df = pd.DataFrame(columns=y_test.columns)

for col in y_train.columns:
        y_train_binary_relevance = y_train[col]
        svclassifier = SVC(kernel='rbf', decision_function_shape='ovr')
        grid_cv = GridSearchCV(svclassifier, tuned_parameters, cv=kf, n_jobs=-1)
        grid_cv.fit(X_train, y_train_binary_relevance)
        best_grid_cv = grid_cv.best_estimator_
        y_pred_train = best_grid_cv.predict(X_train)
        y_pred_test = best_grid_cv.predict(X_test)
        
        y_pred_train_df.loc[:, col] = y_pred_train
        y_pred_test_df.loc[:, col] = y_pred_test

        print("Label ", col)
        print("Best Gaussian kernel Width:", grid_cv.best_params_['gamma'])
        print("Best SVM Penalty:", grid_cv.best_params_['C'])
        
```

Calculating metrics

```python
exact_match_ratio_train, hamming_loss_train = getMetrics(y_train, y_pred_train_df)
exact_match_ratio_test, hamming_loss_test = getMetrics(y_test, y_pred_test_df)

print("\nTrain Exact Match ratio:", exact_match_ratio_train)
print("Test Exact Match ratio:", exact_match_ratio_test)

print("\nTrain Hamming loss:",hamming_loss_train)
print("Test Hamming loss:",hamming_loss_test)
```

Choose k ∈ {1, 2, . . . , 50} automatically based on one of the methods provided in the slides (CH or Gap Statistics or scree plots or Silhouettes) or any other method you know.
(b) In each cluster, determine which family is the majority by reading the true labels. Repeat for genus and species.
(c) Now for each cluster you have a majority label triplet (family, genus, species). Calculate the average Hamming distance, Hamming score, and Hamming loss5 between the true labels and the labels assigned by clusters.

Answers to all the above 3questions are done below together

```python
def majority_label(k, cluster_label, y):
    majority = pd.DataFrame(columns = y.columns)
    for i in range(k):
        idx, = np.where(cluster_label == i)
        cluster_samples = y.iloc[idx, :]
        temp = []
        for label in y.columns:
            major = cluster_samples.loc[:, label].value_counts().index[0]
            temp.append(major)
        majority.loc[i] = temp
    return majority
```

```python
def evaluate_metric(cluster_major, cluster_labels, y):
    miss_clf = 0
    for i in range(len(cluster_major)):
        idx, = np.where(cluster_labels == i)
        for label in y.loc[idx].values:
            miss = (label != cluster_major.loc[i].values)
            miss_clf += np.sum(miss)
    hamming_dist = miss_clf / y.shape[0]
    hamming_loss = miss_clf / (y.shape[0] * y.shape[1])
    return hamming_dist, hamming_loss
```

```python
print("The average Hamming Distance is: ", np.mean(hamming_dist))
print("The standard deviation of the simulation is ", np.std(hamming_dist))
```
